## Compute Area Under Curve for each state

```python
from scipy.integrate import trapz

df = pd.read_csv("./copy_alnspectra_A.dat", sep=" ")

valuesQ0 = df['CountsQ0'].to_numpy()
valuesQ1 = df['CountsQ1'].to_numpy()
valuesQ2 = df['CountsQ2'].to_numpy()

H0 = trapz(valuesQ0, df['Energy'])
H1 = trapz(valuesQ1, df['Energy'])
H2 = trapz(valuesQ2, df['Energy'])

noGammaProb    = H0/(H0+H1+H2)
FirstGammaProb = H1/(H0+H1+H2)
OtherGammaProb = H2/(H0+H1+H2)

print('noGammProb = {:.5f}%'.format(noGammaProb))
print('FirstGammProb = {:.5f}%'.format(FirstGammaProb))
print("OtherGammProb = {:.5f}% (or just second excited state, don't know right now)".format(OtherGammaProb))
print("Sum is actually 100")
```

```python
event = 36

fig = plt.figure()
fig.set_figheight(18.0)
fig.set_figwidth(21.0)

# Initial parameters of the cylinder
r = 168
x = np.float64(candidateInfo[(candidateInfo['event'] == event)]['hit_xf'].values)
y = np.float64(candidateInfo[(candidateInfo['event'] == event)]['hit_zf'].values)
z = np.float64(candidateInfo[(candidateInfo['event'] == event)]['hit_yf'].values)
t = np.float64(candidateInfo[(candidateInfo['event'] == event)]['hit_time'].values)

# Tranversal projection
theta = np.arctan2(y, x)
phi = theta + np.pi
x_proj = r * phi
y_proj = z

# Plot the colobar
ax_cbar = fig.add_subplot(3, 12, 10)
matplotlib.colorbar.ColorbarBase(ax_cbar, cmap='turbo',
                                 norm=matplotlib.colors.Normalize(vmin=np.min(t), vmax=np.max(t)))
ax_cbar.set_title("Time (ns)")

# Plot the upper lid
ax1 = fig.add_subplot(3, 3, 2)
ax1.scatter(x[z > 120], y[z > 120], c=t[z > 120], cmap='turbo')

ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_ylim(-150, 150)
ax1.set_xlim(-150, 150);

# Plot the cylinder 'body'
ax2 = fig.add_subplot(3, 1, 2)
ax2.scatter(x_proj, y_proj, c=t, cmap='turbo')

ax2.set_xlabel('X')
ax2.set_ylabel('Z')
ax2.set_ylim(-120, 120);

# Plot the bottom lid
ax3 = fig.add_subplot(3, 3, 8)
ax3.scatter(x[z < -120], y[z < -120], c=t[z < -120], cmap='turbo')

ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.set_ylim(-150, 150);
ax3.set_xlim(-150, 150);
```

- Cojo el número total de fotones ópticos creados fuera del BGO (i.e. Cherenkov)

Para las gammas:
- Cojo el número total de fotones ópticos creados dentro del BGO (i.e. Scintillation)

La diferencia se debe a que en más de la mitad de los eventos de gamma, la gamma no deposita toda su energía en el BGO y sale fuera, o salen electrones, que provocan Cherenkov. ## Time between first and last Scintillation light hit in every event

```python
time = []
for i in np.unique(df_trueHits_bgo['event_id']):
    # Iterate over all true_hit_parent in event using groupby for fast iteration
    iterator = df_trueHits_bgo[df_trueHits_bgo['event_id'] == i].groupby(['true_hit_parent'])
    
    # Take variables we are interested in
    for jt, t in iterator['true_hit_time']:
        # Save only if ring has more than 5 points
        event  = i
        parent = jt
        hit_t  = [i for i in t.values if i > 0]
        
        # Save
        tmp_times = np.array([i, jt, np.min(hit_t), np.max(hit_t), np.max(hit_t)-np.min(hit_t)], dtype='object')
        time.append(tmp_times)
```

```python
time_diff = [i[-1] for i in time]

plt.hist(time_diff, bins=50, histtype='step');

plt.title('$\Delta$t between first and last true hit from scintillation light');
plt.xlabel('t [ns]');

plt.text(2000, 80, 'Mean $\Delta$t: {:.2f} ns'.format(np.mean(time_diff)));
```
